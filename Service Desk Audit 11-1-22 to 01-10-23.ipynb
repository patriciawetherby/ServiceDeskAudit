```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df = pd.read_csv("General Service Desk Support(1-253).csv")
```

```python
df['Start time'] = pd.to_datetime(df['Start time'])
```

```python
df
```

```python
df['Weekday'] = df['Start time'].dt.dayofweek
```

```python
df
```

```python
df['Weekday'] = df['Weekday'].replace({0:'Tuesday',1:'Wednesday',2:'Thursday',3:'Friday',4:'Monday'})
```

```python
plt.figure(figsize=(10,4),dpi=200)
plt.title('Total Users per Day: 11-1-22 to 01-10-23')
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
sns.countplot(x = "Weekday", data = df, order = day_order);
plt.ylabel('Users Helped')
plt.xlabel('Day of the Week')
plt.savefig('ServiceDeskDays.png')
```

```python
df = df.set_index('Start time')
```

```python
df['Hours'] = df.index.hour
```

```python
df['Hours']
```

```python
df['Hours'] = df['Hours'].replace({7:'7:00am',8:'8:00am',9:'9:00am',10:'10:00am', 11:"11:00am",12:"12:00pm",13:"1:00pm",14:"2:00pm",15:'3:00pm',16:'4:00pm',17:'5:00pm',18:'6:00pm'})
```

```python
plt.figure(figsize=(10,4),dpi=200)
plt.title('Total Users Helped per Hour: 11-1-22 to 01-10-23')
hour_order = ["7:00am", "8:00am", "9:00am", "10:00am", "11:00am","12:00pm","1:00pm","2:00pm",'3:00pm','4:00pm','5:00pm','6:00pm']
sns.countplot(x='Hours', data=df, order=hour_order);
plt.ylabel('Users Helped')
plt.xlabel('Hours(PST)')
plt.savefig('ServiceDeskHours.png')
```
